## Упражнения по библиотеке Numpy

```python
import numpy as np
import itertools
```

**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
# your code
def f1(a):
    a[(a > 3) & (a < 8)] *= -1
    return a
a = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(f1(a))
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
# your code
def f2(a):
    a[a.argmax()] = 0
    return a
a = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(f2(a))
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
# your code
def f3(a):
    result = np.array(list(itertools.product(*a)))
    return result
a = np.array([[1, 2], [3, 4]])
print(f3(a))
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
# your code
def f4(A, B):
    row_mask = np.ones(len(A), dtype = bool)
    for row in B:
        for col in row:
            mask = (A == col).any(axis = 1)
            row_mask = row_mask & mask
    return np.where(row_mask)[0]

A = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [3, 1, 2],
    [1, 2, 4]
])

B = np.array([
    [1, 2],
    [2, 3]
])

print(f4(A, B))
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
# your code
def f5(A):
    A = A[A.min(axis = 1)!= A.max(axis = 1)]
    return A

A = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [7, 7, 7]
])
print(f5(A))
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
# your code
def f6(A):
    A = np.unique(A, axis = 0)
    return A

A = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [1, 2, 3],
    [7, 8, 9],
    [4, 5, 6]
])

print(f6(A))
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# your code
def diag_product(A):
    result = 1
    rows = len(A)
    cols = len(A[0])
    n = min(rows, cols)
    for i in range(n):
        if A[i][i] != 0:
            result *= A[i][i]
    return result

def diag_product_np(A):
    diag = np.diag(A)
    result = np.prod(diag[diag != 0])
    return result

A = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [1, 2, 3],
    [2, 3, 4]
])

print(diag_product(A))
print(diag_product_np(A))
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# your code
def multiset(x, y):
    if len(x) != len(y):
        return False
    d = {}
    for elem in x:
        if elem in d:
            d[elem] += 1
        else:
            d[elem] = 1
    for elem in y:
        if elem not in d or d[elem] == 0:
            return False
        else:
            d[elem] -= 1

    return True

def multiset_np(x, y):
    result = np.array_equal(np.sort(x), np.sort(y))
    return result

x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])

print(multiset(x, y))
print(multiset_np(x, y))
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# your code
def max_elem_np(x):
    good = x[1:][x[:-1] == 0]
    if good.size > 0:
        return np.max(good)
    else:
        return None

def max_elem(x):
    max_val = None
    for i in range(1, len(x)):
        if x[i-1] == 0:
            if max_val is None or x[i] > max_val:
                max_val = x[i]
    return max_val

x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
a = np.array([6, 5, 7, 0])
b = np.array([6, 5, 7])

print(max_elem(x))
print(max_elem_np(x))
print(max_elem(a))
print(max_elem_np(a))
print(max_elem(b))
print(max_elem_np(b))
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# your code
def rle(x):
    if len(x) == 0:
        return np.array([]), np.array([])
    keys = [x[0]]
    counts = []
    count = 1

    for i in range(1, len(x)):
        if x[i] == x[i-1]:
            count += 1
        else:
            counts.append(count)
            keys.append(x[i])
            count = 1
    counts.append(count)

    return np.array(keys), np.array(counts)

def rle_np(x):
    if x.size == 0:
        return np.array([]), np.array([])
    change_idx = np.where(x[1:] != x[:-1])[0] + 1
    sections = np.concatenate(([0], change_idx, [len(x)]))
    values = x[sections[:-1]]
    counts = np.diff(sections)

    return values, counts

x = np.array([2, 2, 2, 3, 3, 3, 5])
a = np.array([])
print(rle(x))
print(rle_np(x))
print(rle(a))
print(rle_np(a))
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
# your code
import math
from scipy.spatial.distance import cdist
import time

def dist(X, Y):
    result = []
    
    for x in X:
        row = []
        for y in Y:
            d = math.sqrt(sum((a - b) ** 2 for a, b in zip(x, y)))
            row.append(d)
        result.append(row)
    return result

def dist_np(X, Y):
    diff = X[:, np.newaxis] - Y
    result = np.linalg.norm(diff, axis=-1)
    return result

X_data = np.random.rand(1000, 10)
Y_data = np.random.rand(1000, 10)
X_list = X_data.tolist()
Y_list = Y_data.tolist()

start = time.time()
res_py = dist(X_list, Y_list)
print(f"Python: {time.time() - start:.4f}s")

start = time.time()
res_np = dist_np(X_data, Y_data)
print(f"NumPy: {time.time() - start:.4f}s")

start = time.time()
res_scipy = cdist(X_data, Y_data)
print(f"SciPy: {time.time() - start:.4f}s")
```
